```python
x = Weiner()

df = x.brownian_motion(100,5)
fig, ax = plt.subplots()
df.plot(ax=ax)
```

```python
x1 = Weiner(initial_condition=0,generator=stats.norm, mu=0.2, kw_generator={'scale':1},freq_mu='A', seed=9)

df1 = x1.brownian_motion(50*12,10, freq='M')
df1.plot()
```

```python
x2 = Weiner(initial_condition=0,generator=stats.norm, mu=0.2/12, kw_generator={'scale':0.2887},freq_mu='M', seed=9)

fig,ax = plt.subplots()
df2 = x2.brownian_motion(50*12,10, freq='M')
df2.plot(ax=ax)
df2a = x2.brownian_motion(50*12,3, freq='M',interval=0.1)
df2a.plot(ax=ax)
```

```python
x3 = Weiner(initial_condition=10,generator=stats.norm, mu=0.03, kw_generator={'scale':0.26},freq_mu='A', seed=9)

df3 = x3.geometric_brownian_motion(20*12,20, freq='M')
df3.plot()
```
